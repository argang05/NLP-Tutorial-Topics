```python
#  Load or create a blank NLP model for English
nlp = spacy.blank("en")
```

```python
#Training the ner entity for new labels:
if 'ner' not in nlp.pipe_names:
    ner = nlp.add_pipe('ner')
else:
    ner = nlp.get_pipe('ner')

#  Add entity labels from training data to the NER pipeline
for _, annotations in train_data:
    for ent in annotations['entities']:
        if ent[2] not in ner.labels:
            ner.add_label(ent[2])  # Add FOOD, ORDER_ID, PHONE labels

#  Disable other pipes to focus only on training NER
other_pipes = [pipe for pipe in nlp.pipe_names if pipe != 'ner']
with nlp.disable_pipes(*other_pipes):

    #  Initialize training optimizer
    nlp.initialize()

    #  Define number of training epochs
    epochs = 50  

    for epoch in range(epochs):
        random.shuffle(train_data)  # Shuffle training data each epoch
        losses = {}  # Track training losses

        #  Process data in small batches to optimize training
        batches = minibatch(train_data, size=5)
        for batch in batches:
            examples = []
            for text, annotations in batch:
                doc = nlp.make_doc(text)  # Create a blank spaCy Doc
                example = Example.from_dict(doc, annotations)  # Convert annotations to spaCy format
                examples.append(example)

            #  Update the model with examples and track losses
            nlp.update(examples, drop=0.5, losses=losses)

        print(f"Epoch: {epoch + 1}, Losses: {losses}")  # Print loss per epoch

#  Save the trained model for future use
nlp.to_disk("custom_ner_model")
print("Training Complete! Model saved as 'custom_ner_model'")
```
